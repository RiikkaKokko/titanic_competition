```python
import pandas as pd
import numpy as np
```

```python
#reading the csv-file
df=pd.read_csv("train.csv")
```

```python
#kõikidele tühjadele vanuseridadele arvutame ja lisame vanuste keskmise väärtuse
df['Age'] = df['Age'].fillna((df['Age'].mean()))
#kõikidele tühjadele sadamaridadele lisame eelneva rea väärtuse
df["Embarked"].fillna( method ='ffill', inplace = True) 
#kõikidele tühjadele cabin ridadele lisame väärtuse "empty"
df["Cabin"].fillna("Empty", inplace = True)
```

```python
#teeme uue muutuja "df1" ning splittime nimest eesnime ja perekonnanime eraldi
df1 = pd.DataFrame(df.Name.str.split(' ',1).tolist(),
                                   columns = ['First','Last'])
```

```python
#vaatame andmeid
df1.head()
```

```python
#splitime eraldi eesliite
df1= pd.DataFrame(df1.Last.str.split(' ',1).tolist(),
                                   columns = ['prefix','Last'])
```

```python

#vaatame andmeid, kas tulemus on õige
df1.head()
```

```python
#lisame eesliite eraldi veergu, muutujasse "df"
df['prefix']=df1.prefix
```

```python
#vaatame andmeid, kas tulemus on õige
df.head()
```

```python
#kustutame veerud "name", "ticket", "cabin" ja "PassengerID", need veerud ei anna meile piisavalt unikaalset informatsiooni. 
df=df.drop(columns=["Name"])
df=df.drop(columns=["Ticket"])
df=df.drop(columns=["Cabin"])
df=df.drop(columns=["PassengerId"])
```
